```python
import requests

# Test 1: Simple fetch (like v5 which worked)
url1 = 'https://data.lacity.org/resource/d5tf-ez2w.json?$limit=100'
r1 = requests.get(url1, timeout=60)
r1.raise_for_status()
print('Test 1 OK:', len(r1.json()), 'records')

# Test 2: Add $where clause
url2 = 'https://data.lacity.org/resource/d5tf-ez2w.json?$limit=100&$where=date_occ%3E%272022-01-01T00:00:00%27'
r2 = requests.get(url2, timeout=60)
r2.raise_for_status()
print('Test 2 OK:', len(r2.json()), 'records')

# Test 3: Add $select
url3 = 'https://data.lacity.org/resource/d5tf-ez2w.json?$limit=100&$where=date_occ%3E%272022-01-01T00:00:00%27&$select=dr_no,date_occ,lat,lon'
r3 = requests.get(url3, timeout=60)
r3.raise_for_status()
print('Test 3 OK:', len(r3.json()), 'records, keys:', list(r3.json()[0].keys()))

# Test 4: 10K records
url4 = 'https://data.lacity.org/resource/d5tf-ez2w.json?$limit=10000&$where=date_occ%3E%272022-01-01T00:00:00%27&$select=dr_no,date_occ,lat,lon'
r4 = requests.get(url4, timeout=120)
r4.raise_for_status()
print('Test 4 OK:', len(r4.json()), 'records')

print('ALL URL TESTS PASSED')
```
